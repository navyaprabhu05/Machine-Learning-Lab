```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your file (replace filename.csv with actual name)
df = pd.read_csv('indian_food.csv')

# Display first 5 rows
df.head()
```

```python
df.shape                # Rows and columns
df.columns              # Column names
df.info()               # Data types and non-null counts
df.describe()           # Summary stats for numerical columns
df.dtypes               # Check data types
df.isnull().sum()       # Count of missing values per column
```

**Top Rated Dishes**

```python
df[df['avg_rating'] >= 4.5].head()
```

**Sort Dishes by Popularity**

```python
df.sort_values(by='orders_num', ascending=False).head()
```

**Average Rating per State**

```python
df.groupby('state')['avg_rating'].mean().sort_values(ascending=False)
```

Average Rating per course

```python
course_rating = df.groupby('course')['avg_rating'].mean()
course_rating.plot(kind='bar', title='Avg Rating by Course', color='skyblue')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

**Histogram of prep times**

```python
df['prep_time'].hist(bins=20, color='orange')
plt.title('Distribution of Prep Time')
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()
```
